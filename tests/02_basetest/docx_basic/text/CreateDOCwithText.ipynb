```python
import os, sys, ipytest
from pathlib import Path
ipytest.autoconfig()
```

```python
REPORT_DIR = Path(os.getcwd())
#print(os.getcwd(),REPORT_DIR)

TEST_ROOT = REPORT_DIR.parent.parent.parent
if str(TEST_ROOT) not in sys.path:
    sys.path.append(str(TEST_ROOT))

from baseTestRoot import *

import Scriptum # type: ignore
mydoc = None
```

```python
def base_setup(tmp_path):

    setupTestEnvironment(tmp_path, DATA_SOURCE, REPORT_DIR, 
                         include_patterns = [
                            "*.rdf",
                            "template_text.docx",
                         ])
    os.chdir(tmp_path)
    print(tmp_path)
```

```python
%%ipytest -vvv -s
def test_me(tmp_path):
    global mydoc
    base_setup(tmp_path)

    rdf = Scriptum.ReportDataFile('word_text.rdf')
    if rdf.errors: 
        # assemble collected errors if any
        print('\n'.join(rdf.errors))
    #rdf.inspect()
    #rdf.showFiles()

    mydoc = Scriptum.ManagedDocx('template_text.docx')
    print(mydoc.errors, mydoc.warnings)

    mydoc.typesetting(rdf,addcopy=True, directfill=True, globalfill=True, cleanup=True, removetemplate=True, cleardust=True, setproperties=True)

    mydoc.save('result_word.docx')#, finish=True, createpdf=True)

    print(tmp_path)
```
