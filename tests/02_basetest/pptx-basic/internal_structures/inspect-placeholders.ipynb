# Inspect placeholder shapes
in order to run the following code you should:
- have package python-pptx installed

```python
from pptx import Presentation
from pptx.util import Inches, Pt, Cm
```

```python
document = Presentation('placeholders.pptx')
```

```python
for layout in document.slide_layouts:
    print(layout.name)
baselayout = document.slide_layouts[0]
```

```python
newslide=document.slides.add_slide(baselayout)
```

```python
for ph in baselayout.placeholders:
    print(ph.text, ph.shape_id, ph.placeholder_format.type)#ph._element.getchildren()[0].getchildren()[-1].getchildren()[0].xml)

ph = baselayout.placeholders[4]
#print(ph.text, ph._parent, ph._element.getchildren()[0].getchildren()[-1].xml)
ch = ph._element.getchildren()[0].getchildren()[-1].getchildren()[0]
ch.attrib
```

```python
for ph in newslide.placeholders:
    print(ph, ph.placeholder_format.type, ph.shape_id, dir(ph))
```
