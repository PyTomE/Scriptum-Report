# Inspect shapes
in order to run the following code you should:
- have package python-pptx installed

```python
import pptx
print(pptx.__version__)
from pptx import Presentation
from pptx.util import Inches, Pt, Cm
```

```python
document = Presentation('shapes.pptx')
```

```python
slide = document.slides[0]
```

```python
for sh in slide.part.slide.shapes:
    print(sh,sh.shape_type)
    if hasattr(sh,'shapes'):
        for subsh in sh.shapes:
            print('  ',subsh,subsh.shape_type)
    if sh.has_text_frame:
        print(sh.text)
```
