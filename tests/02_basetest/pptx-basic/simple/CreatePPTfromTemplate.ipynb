```python
import os, sys, ipytest
from pathlib import Path
ipytest.autoconfig()
```

```python
REPORT_DIR = Path(os.getcwd())
#print(os.getcwd(),REPORT_DIR)

TEST_ROOT = REPORT_DIR.parent.parent.parent
if str(TEST_ROOT) not in sys.path:
    sys.path.append(str(TEST_ROOT))

from tests.baseTestRoot import *

import Scriptum # type: ignore
myppt = None
```

```python
def base_setup(tmp_path):

    setupTestEnvironment(tmp_path, DATA_SOURCE, REPORT_DIR, 
                         include_patterns = [
                            "*.rdf",
                            "template.pptx",
                         ])
    os.chdir(tmp_path)
    print(tmp_path)
```

```python
%%ipytest -vvv -s
def test_me(tmp_path):
    global myppt
    base_setup(tmp_path)

    rdf = Scriptum.ReportDataFile('powerpoint_simple.rdf')
    if rdf.errors: 
        # assemble collected errors if any
        print('\n'.join(rdf.errors))
    #rdf.inspect()
    #rdf.showFiles()

    myppt = Scriptum.ManagedPptx('template.pptx')
    print(myppt.errors, myppt.warnings)

    myppt.artist(rdf, directfill=True, globalfill=True, cleardust=True, setproperties=True)

    myppt.document.core_properties.title='AutoReport'
    myppt.remove_slide(0)
    
    myppt.save('result_ppt.pptx')#, finish=True, createpdf=True)

    print(tmp_path)
```
